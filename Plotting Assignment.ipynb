```python
# Depicts the distribution of CO2
co2 = fuel_econ['co2']
plt.figure(figsize=(10, 6))
plt.hist(co2, bins=10, edgecolor='white')
plt.xlabel('CO2 Emissions (g/km)')
plt.ylabel('Number of vehicles')
plt.title('Distribution of CO2 Emissions')
plt.show
```

```python
# Experimenting with bin size 20
co2 = fuel_econ['co2']
plt.figure(figsize=(10, 6))
plt.hist(co2, bins=20, edgecolor='white')
plt.xlabel('CO2 Emissions (g/km)')
plt.ylabel('Number of vehicles')
plt.title('Distribution of CO2 Emissions')
plt.show
```

When using bins=10, each bar in the histogram is wider compared to bins=20 because fewer bins cover the same data range. The bin width controls how the data is grouped along the x-axis  wider bins group more data points together, while narrower bins provide more detail by dividing the data into smaller intervals.

```python
# Depicts the distribution of combined fuel
co2 = fuel_econ['comb']
plt.figure(figsize=(10, 6))
plt.hist(co2, bins=10, edgecolor='white')
plt.xlabel('Combination of fuels')
plt.ylabel('Number of vehicles')
plt.title('Distribution of Combined Fuel (comb)')
plt.show
```

```python
# Depicts the distribution of cylinder
co2 = fuel_econ['comb']
plt.figure(figsize=(10, 6))
plt.hist(co2, bins=10, edgecolor='white')
plt.xlabel('Number of  cylinders')
plt.ylabel('Number of vehicles')
plt.title('Distribution of Cylinders in Vehicles')
plt.show
```

```python
numerical_cols = fuel_econ.select_dtypes(include='number')
correlation_matrix = numerical_cols.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=False, fmt='.2f',cmap="coolwarm", cbar=True)
plt.title('Correlation Heatmap of Vehicles and fuel efficiency')
plt.show()
```

**Interpretation of the heatmap**

1. Strong positive correlation (close to +1 represented by red tones): meaning 2 variables increase together.
  E.g: cylinders and displ; as there is more cylinders the displacement become larger.

2. Strong negative (close to -1 represented by blue tones): meaning as one increase the other decreases.
   E.g: CO2 and ghg score; as there is more carbon dioxide the green house score decreases.

3. zero or near colloration(grey tones): No clear linear relationship between the two variables.

**Story told by the visualization:**

-More fuel efficient vehicles release less carbon dioxide; combined fuel efficiency (comb) increases and the vehicle release less  CO2.

-Larger engines tend to emit more CO2 as there is a strong positive correlation between displ,cylinders and co2.

-As there is a high fuel sxore there is a lower green house impact
